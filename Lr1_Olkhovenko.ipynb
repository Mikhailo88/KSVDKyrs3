```python
# Обернена
a = np.matrix([[1, -3], [2, 5]])

a_inv = np.linalg.inv(a)
a_inv
```

```python
a = np.array([[2, 8, 1], [9, 5, 6]])
b = np.array([[7, 6, 2], [10, 9, 12]])

c1 =  a*b
print("Поелементний добуток:")
print(c1)
```

```python
a = np.array([[0, -3, 4],
              [-2, 0, 1],
              [3, 1, 0]])


eigenvalues, eigenvectors = np.linalg.eig(a)
```

```python
print("V:")
print(eigenvalues)
```

```python
print("\nW:")
print(eigenvectors)
```
