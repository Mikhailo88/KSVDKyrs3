# Завдання 3

```python
import altair as alt
from vega_datasets import data
import plotly.express as px
```

1. Горизонтальна діаграма про врожайність пшениці

```python
df = data.wheat()

bars = alt.Chart(df).mark_bar().encode(
    x=alt.X('wheat:Q', title='Врожайність пшениці'),
    y=alt.Y('year:O', title='Рік')
)

text = bars.mark_text(
    align='left',
    baseline='middle',
    dx=3
).encode(
    text='wheat:Q'
)

chart = bars + text
chart
```

2. Багатошарова діаграма про виробництво електроенергії в штаті Айова

```python
import altair as alt
from vega_datasets import data

df = data.iowa_electricity()

chart2 = alt.Chart(df).mark_bar(opacity=0.7).encode(
    x=alt.X('year:O', title='Рік'),
    y=alt.Y('net_generation:Q', title='Чисте виробництво електроенергії'),
    color=alt.Color('source:N', title='Джерело енергії')
)

chart2
```

3. Графік для ірисів

```python
df = px.data.iris()

# scatter plot
chart3 = px.scatter(
    df,
    x="sepal_width",          # ширина чашелистика
    y="sepal_length",         # довжина чашелистика
    size="petal_length",      # розмір точок — довжина пелюстки
    color="species",          # колір — вид
    hover_data=["petal_width"]  # додаткові дані при наведенні
)

chart3
```
